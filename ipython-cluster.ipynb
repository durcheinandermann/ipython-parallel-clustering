##Clustering
Clustering refers to a set of methods of unsupervised learning which aim to find groups in a data set. In general, these groups are desired to be homogenous in themselves but heterogeneous from each other. What exactly defines a homogeneous group with "similar" data points oftentimes depends on the domain of the application but in practise, the distance between features is often employed. [ref ISL p. 385].
During clustering, every data point is assigned to one group. [Clustering can also be used to identify groups of similar features, ISL p. 386]

#K-means
K-means clustering is an iterative clustering technique and constitutes one of the most popular methods in clustering. It relies on a predetermined input k which denotes the number of clusters to be found. Various strategies for choosing k exist but are not subject of this discussion. K-means then finds K non-overlapping clusters and hard-assigns each data point to exactly one cluster.
The criterion to measure the quality of a given clustering is the sum of squared distances between data points and corresponding centroids, which reflects the overall goal of finding homogeneous groups of similar points. The lower this criterion, the more homogeneous the clusters and the better the fit.

The algorithm works in two steps and finds a local optimum of the before-mentioned criterion:
    0. Pick k points as initial cluster centroids.
    1. Assign data points to the nearest centroid (cluster). "Nearest" is frequently measured in terms of squared Euclidean distance between a data point and the centroids.
    2. Update the cluster centroids by calculating a vector of means for all features across all data points in each cluster.

    Steps 1 and 2 are repeated until convergence, that is until the assignments of data points no longer change or the cluster centers remain the same.

#K-means/initializations
A common strategy for initializing cluster centroids is random selection of points. This approach can be motivated by the fact that cluster centroids tend to be in "dense" areas where many points reside; thus choosing centers randomly will yield points close to true cluster with a higher probability. On the other hand, it may very well happen that two such randomly chosen points are very close to the same center - something we do not desire. Moreover, depending on this selection, the outcomes of K-means clustering may differ as the algorithm will find a local optimum starting from the initlization.
Arthur and Vassilvitskii [XXX] have shown that a smarter initliazation strategy can improve this by achieving a lower value of the above mentioned optimization criterion - even frequently finding the optimal result on synthetic data - and also decrease the number of iterations [Arthur et al., p. 9]. Their approach is termed K-means++ and only chooses the first centroid at random. Sequentially, k-1 additional centers are chosen with a probability proportional to their squared distance from the previously selected centroids. This ensures diverse centroids and avoids choosing outliers. [Marco CapÃ³ et al., p. 930] One drawback of K-means++ is that its initialization needs k passes over the data, which consumes a lot of time on very big data sets. To overcome this problem, Bahmani et al. ["Scalable K-means"] developed a scalable version of k-means++ named k-means||.

Ipython parallel

IPython parallel is a parallelization framework for IPython. It not only allows parallelization by harvesting the power of multiple CPUs or GPUs on the local machine but also allows for distributed computing across various machines which may reside at remote locations. The use of IPyhton Parallel requires the installation of the "ipyparallel" package.

IPython parallel's architecture constis of a client, a controller and a set of engines. The controller consists of a hub and a scheduler. The hub keeps track of what's going on on the cluster, while the schedulers provide an asyncrhonous, in other words non-blocking during busy engines, interface between client and engines. Clients and engines connect to the controller.
The engines listen for requests, run code and return the results.

The work to be done requested by the client can be sent to different sets of engines through so-called "views". Various modes exist to manually control assimgent of work to specific engines or to allow for a load-balanced assigment determined by the scheduler. Moreover, users can state whether execution should be syncrhonous or asynchrononous, that is waiting or no waiting for results before continuing.

Means of parallelization

Common methods are the use of a parallel equivalent of Python's map function as well as executing functions remotely through ipyparallel's apply method. However, using IPyhton Parallel's "magic" commands like %px make it possible to execute single lines of code directly on the engines, which for example allows quick implementation of powerful list comprehension commands.

Note that the engines have their own namespaces. Thus, to allow computations on engines, the engines must be equipped with the same functions / packages and have access to the necessary variables/data. The data can either be pushed to engines as a whole or partitioned by scattering it across engines. Using the latter approach, data can be re-gathered after computations on engines have finished.

One drawback of a framework like IPython Parallel is the overhead cost of sending data back and forth (over the network) as well as the coordination of client, schedulers and engines. Thus, the question arises as to under which circumstances the benefits of parallel computation surpass the initial overhead costs.
```python
from sklearn.datasets.samples_generator import make_blobs

def create_data(size_list, k, features=2):
    res = []
    for number in size_list:
        np.random.seed(seed=None)
        a, b = make_blobs(n_samples=number, centers=k, n_features=features)
        res.append(a)
    return res
```

```python
def init_centres(X, k, method):
    if method=="random":
        np.random.seed(seed=604)
        return X[np.random.choice(X.shape[0],k,replace=False),:]
    if method=="k++":
        #pick first point at random
        np.random.seed(seed=604)
        res = X[np.random.choice(X.shape[0],1,replace=False),:]
        #choose subsequent points
        if k>1:
            for j in range(1,k):
                #calculate distances
                Distances = np.array([min([np.linalg.norm(x-p)**2 \
                                        for p in res]) for x in X])
                #model distribution
                pdf = Distances/Distances.sum()
                cdf = pdf.cumsum()
                #choose point
                r = np.random.random()
                choice = np.where(cdf >= r)[0][0]
                #add to result array
                res = np.vstack((res, X[choice]))
            return res
        else:
            return res
```

```python
def sk_means(X, k, seeding="random", ReturnPlot = True): #serial k-means
    if seeding=="random":
        c = init_centres(X, k, method="random")
    else:
        c = init_centres(X, k, method="k++")
    a = np.ndarray([len(X),1], )
    a = a.ravel()
    while True:
        #Assign points to clusters
        for q in range(0, len(a)):
            a[q] = np.argmin([np.linalg.norm(x - X[q]) for x in c])
        #Re-calculate cluster centres
        c_old = c.copy()
        for i in range(0,len(c)):
            c[i,0] = np.mean(X[a==i,0])
            c[i,1] = np.mean(X[a==i,1])
        if np.array_equal(c, c_old):
            if ReturnPlot == True:
                Fig1, ax = plt.subplots()
                for i in range(0,k):
                    ax.plot(X[a==i,0], X[a==i,1], '.')
                ax.plot(c[:,0], c[:,1], 'k^')
                return Fig1
            else:
                return a, c
```

```python
def pk_means(X, k, view, seeding="random", ReturnPlot = True, ReturnParam = False):
    view.activate()
    if seeding=="random":
        c = init_centres(X, k, method="random")
    else:
        c = init_centres(X, k, method="k++")
    a = np.ndarray([len(X),1], )
    a = a.ravel()
    #partition a (array with cluster indices for each point) across engines
    view.scatter('b', a)
    #object b is non-writable for some reason
    %px a = numpy.ndarray([len(b),1], )
    %px a = a.ravel()
    %px del b
    #partition X (array with points) across engines
    view.scatter('X', X) 
    while True:
        #Assign points to clusters
            #scatter data:
            #push c with cluster centres to all engines
            view['c'] = c
            #identify closest clusters for points remotely on partitioned data
            %px for q in range(0, len(a)): a[q] = \
                numpy.argmin([numpy.linalg.norm(x - X[q]) for x in c])
            #gather data
            a = view.gather('a')#, block=True)

        #Re-calculate cluster centres
            c_old = c.copy()
            for i in range(0,len(c)):
                c[i,0] = np.mean(X[a==i,0])
                c[i,1] = np.mean(X[a==i,1])
            #check if cluster centers moved
            if np.array_equal(c, c_old):
                %px del X, a, c
                if ReturnPlot == True:
                    Fig1, ax = plt.subplots()
                    for i in range(0,k):
                        ax.plot(X[a==i,0], X[a==i,1], '.')
                    ax.plot(c[:,0], c[:,1], 'k^')
                    return Fig1
                if ReturnParam == True: #to get parameters for GMM
                    #centroids, covariance matrices, shares of points in each cluster
                    return c, \
                        np.array([np.cov(X[a==i,], rowvar=False) for i in range(0,k)]), \
                        np.array([sum(a==i)/len(a) for i in range(0,k)])
                else:
                    return a, c
```

# Gaussian Mixture Model

Mixture models overcome one of the weaknesses of K-means clustering. In K-means, every point is assigned to exactly one cluster. However, especially when dealing with outliers, which may in trugh not belong to any cluster at all, points in a cluster may become very, very different from each other and clusters are distorted. This is also due to the squared measure of Euclidean distance in K-means, which corresponds to searching for circular clusters where in reality the shapes may be significantly different from circles. [B 427]

Instead of hard-assigning each point to exactly one cluster, mixture models do a soft-assignment, where data points are assigned to clusters along with a probability for the point to truly belong to the corresponding cluster. Gaussian mixture models assume a mixture of Gaussian distributions, each equipped with mean, covariance matrix. The individual Gaussians are linearly combined through mixing parameters assigned to each Gaussian that add up to 1.

Initially, the assignment of points and the distribution parameters are unknown. A commonly employed solution for solving this problem is an iterative approach though the Expectation Maximization algorithm. Similarly to k-means, it consists of two main phases.

    0. initialize parameters through (educated) guesses
    1. Expectation phase: Soft-assign each point to each Gaussian by assigning each point a set of "responsibilities". These reflect the posterior probabilities that a point stems from the corresponding Gaussians given their current distribution parameters.
    2. Maximization phase: update the parameters of each Gaussian through maximum likelihood estimates.

The goodness of the fit is determined by the value of the log-likelihood function. Note that a problem arises if one of the points falls exactly onto one of the centroids; this will cause this point contribute an infinite term to the likelihood and the algorithm will diverge. A possible solution is to regularize the covariance matrices artificially.

Repeat stepts 1 and 2 until convergence, that is once the parameters or the log-likelihood no longer change.  However, it may be useful to limit the number of iterations or set a threshold so that a result is accepted once the log-likelihood or the parameters only marginally change or convergence may take very long. [B 437]Like in the case of K-means, it is not guaranteed that the Expectation Maximization algorithm will find the global optimum. [B 438]

#Initializations
Similarly to K-means, one possibility of initializing the parameters of the Gaussians is to randomly assign points as centroids, use identity matrices for the covariance matrices and set the mixing proportions all equal.

However, the Gaussian mixture model on average requires many more iterations than K-means to converge and every iteration is computationally much more expensive. Thus, it may make sense to first conduct a K-means clustering and use its result as an educated guess for initializing the Gaussians. While the obtained centroids are reused in this scenario, the covariance matrices can be initialized through the sample covariances of the K-means clusters and the mixing proportions may be set equal to the proportion of data points in the corresponding K-means cluster. [B 438]

Like in the K-means case, various techniques exist to estimate the number of distributions, however this is beyond the scope of this discussion.

```python
def sGMM(X, k, threshold = 0, seeding = "random", ReturnPlots = True):
    #list of log likelihood
    likelihood = []
    ###
    #initialize parameters
    ###
    if seeding == "random":
        mixprop = np.ndarray([k,1],)
        mixprop = mixprop.ravel()
        mixprop[:] = 1/k

        mu = X[np.random.choice(X.shape[0],k,replace=False),:]

        #covariance matrix for sigma
        dimensions = np.shape(X)[1]
        sigma = [np.eye(dimensions)] * k

    if seeding == "kmeans++":
        mu, sigma, mixprop = pk_means(X, k, view, seeding="kmeans++", \
                                      ReturnPlot = False, ReturnParam = True)
        dimensions = np.shape(X)[1]
    if seeding == "kmeans":
        mu, sigma, mixprop = pk_means(X, k, view, seeding="random", \
                                      ReturnPlot = False, ReturnParam = True)
        dimensions = np.shape(X)[1]
    
    #responsibility array
    respon = np.zeros([len(X),k])
    
    #count iterations
    itero = 0
    while True:
        itero = itero + 1
        ###
        #Expectation
        ###
        #denominator for expectation calculation
        den = numpy.array([[mixprop[b]*multivariate_normal.pdf(row, \
                    mean= mu[b], cov=sigma[b]) for b in range(0,k)] \
                    for row in X]).sum(axis = 1)
        #calculate responsibility matrix
        respon = numpy.array([[mixprop[b]*multivariate_normal.pdf(row[1], \
                    mean= mu[b], cov=sigma[b]) / den[row[0]] for b in range(0,k)] \
                    for row in enumerate(X)])             
                
        ###
        #Maximization
        ###
        #remember old parameters
        mu_old = mu.copy()
        sigma_old = sigma.copy()
        mixprop_old = mixprop.copy()
        #calculate number of points in each component
        N_ks = np.sum(respon, axis = 0)
        for j in range(0,k):
            #means mu
            mu[j] = 1. / N_ks[j] * np.sum(respon[:, j] * X.T, axis = 1).T
            x_mu = np.matrix(X - mu[j])
            
            #sigma
            sigma[j] = np.array(1 / N_ks[j] * np.dot(np.multiply(x_mu.T, \
                                                        respon[:, j]), x_mu))
            
            #mixprop
            mixprop[j] = 1. / len(X) * N_ks[j]
        
        #record log likelihood
        likelihood.append(sum(np.log(den)))
        
        #check for convergence
        if threshold > 0 and itero > 1:
            if (likelihood[-1] - likelihood[-2]) < threshold:
                if ReturnPlots == True:
                    Fig0, ax = plt.subplots()
                    ax.plot(range(1,itero+1),likelihood)
                    if dimensions==2:
                        Fig1, ax = plt.subplots()
                        for j in range(0,k):
                            ax.scatter(X[np.argmax(respon, axis=1)==j,0], \
                                       X[np.argmax(respon, axis=1)==j,1], 
                                s=np.max(respon[np.argmax(respon, axis=1)==j], \
                                         axis=1)*10)
                        ax.plot(mu[:,0], mu[:,1], 'k^');
                    return Fig1, Fig0
                else:
                    return mu, sigma, mixprop, respon
        else:
            if np.array_equal(mu_old, mu) and np.array_equal(mixprop_old, mixprop) \
                                        and np.array_equal(sigma_old, sigma):
                if ReturnPlots == True:
                    Fig0, ax = plt.subplots()
                    ax.plot(range(1,itero+1),likelihood)
                    if dimensions==2:
                        Fig1, ax = plt.subplots()
                        for j in range(0,k):
                            ax.scatter(X[np.argmax(respon, axis=1)==j,0], \
                                       X[np.argmax(respon, axis=1)==j,1], \
                                s=np.max(respon[np.argmax(respon, axis=1)==j], \
                                        axis=1)*10)
                        ax.plot(mu[:,0], mu[:,1], 'k^');
                    return Fig1, Fig0
                else:
                    return mu, sigma, mixprop, respon
```

```python
###############################################################
#Speed experiment: K-means++, serial vs parallel, different ks#
###############################################################

from matplotlib.offsetbox import AnchoredText
dataset_size = np.arange(500,30000,1000)

#Average over many random datasets:
rloops = 250

#######
# k=2 #
#######
rk = 2

for u in range(1,rloops+1):
    XX = create_data(dataset_size, rk)
    lds = len(dataset_size)

    #RECORD TIMINGS
    ts2 = np.zeros([lds,3],)
    ts2[:,0]=dataset_size

    #serial k means plus plus
    for i in range(0,lds):
        print("s++",u, len(XX[i]))
        z = %timeit -n1 -r1 -o sk_means(XX[i], rk, seeding="k++",\
                                        ReturnPlot = False)
        ts2[i,1] = z.best

    #parallel k means plus plus
    for i in range(0,lds):
        print("p++",u, len(XX[i]))
        z = %timeit -n1 -r1 -o pk_means(XX[i], rk, e_all, seeding="k++",\
                                        ReturnPlot = False)
        ts2[i,2] = z.best
        
    if u == 1:
        tst2 = ts2
    else:
        tst2 = np.dstack((tst2,ts2))

#calculate mean values
ts2 = numpy.mean(tst2, axis=2)

#######
# k=3 #
#######
rk = 3

for u in range(1,rloops+1):
    XX = create_data(dataset_size, rk)
    lds = len(dataset_size)

    #RECORD TIMINGS
    ts3 = np.zeros([lds,3],)
    ts3[:,0]=dataset_size

    #serial k means plus plus
    for i in range(0,lds):
        print("s++",u, len(XX[i]))
        z = %timeit -n1 -r1 -o sk_means(XX[i], rk, seeding="k++",\
                                        ReturnPlot = False)
        ts3[i,1] = z.best

    #parallel k means plus plus
    for i in range(0,lds):
        print("p++",u, len(XX[i]))
        z = %timeit -n1 -r1 -o pk_means(XX[i], rk, e_all, seeding="k++",\
                                        ReturnPlot = False)
        ts3[i,2] = z.best
        
    if u == 1:
        tst3 = ts3
    else:
        tst3 = np.dstack((tst3,ts3))

#calculate mean values
ts3 = numpy.mean(tst3, axis=2)

#######
# k=4 #
#######
rk = 4

for u in range(1,rloops+1):
    XX = create_data(dataset_size, rk)
    lds = len(dataset_size)

    #RECORD TIMINGS
    ts4 = np.zeros([lds,3],)
    ts4[:,0]=dataset_size

    #serial k means plus plus
    for i in range(0,lds):
        print("s++",u, len(XX[i]))
        z = %timeit -n1 -r1 -o sk_means(XX[i], rk, seeding="k++",\
                                        ReturnPlot = False)
        ts4[i,1] = z.best

    #parallel k means plus plus
    for i in range(0,lds):
        print("p++",u, len(XX[i]))
        z = %timeit -n1 -r1 -o pk_means(XX[i], rk, e_all, seeding="k++",\
                                        ReturnPlot = False)
        ts4[i,2] = z.best
        
    if u == 1:
        tst4 = ts4
    else:
        tst4 = np.dstack((tst4,ts4))

#calculate mean values
ts4 = numpy.mean(tst4, axis=2)

################
#     plot     #
################
f_skmeans_vs_pkmeans, (ax1, ax2, ax3) = plt.subplots(3, figsize=(8,11),\
                                                sharex=True, sharey=True)

#ax1 for plot where k=2
ax1.plot(ts2[:,0], ts2[:,1], 'r--', label="Serial execution K-means++")
ax1.plot(ts2[:,0], ts2[:,2], 'b--', label="Parallel execution K-means++")

#ax2 for plot where k=3
ax2.plot(ts3[:,0], ts3[:,1], 'r--', label="Serial execution K-means++")
ax2.plot(ts3[:,0], ts3[:,2], 'b--', label="Parallel execution K-means++")

#ax3 for plot where k=4
ax3.plot(ts4[:,0], ts4[:,1], 'r--', label="Serial execution K-means++")
ax3.plot(ts4[:,0], ts4[:,2], 'b--', label="Parallel execution K-means++")

# Fine-tune figure; make subplots close to each other and hide x ticks for
# all but bottom plot.
f_skmeans_vs_pkmeans.subplots_adjust(hspace=0)
plt.setp([a.get_xticklabels() for a in f_skmeans_vs_pkmeans.axes[:-1]],\
                                                 visible=False)

#add labels etc.
ax3.set_xlabel("data set size")
ax2.set_ylabel("execution time (s)")

ax1.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3,
           ncol=2, mode="expand", borderaxespad=0.)

ax1.add_artist(AnchoredText("k=2", loc=2))
ax2.add_artist(AnchoredText("k=3", loc=2))
ax3.add_artist(AnchoredText("k=4", loc=2))

f_skmeans_vs_pkmeans.savefig("f_skmeans_vs_pkmeans.eps", format="eps",\
                             bbox_inches="tight")
f_skmeans_vs_pkmeans.savefig("f_skmeans_vs_pkmeans.jpg", format="jpg",\
                             bbox_inches="tight")
```
